```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
X = np.random.randn(10,4)
X
```

```python
W1 = np.random.randn(4, 3)/100
W1.round(3)
```

```python
b1 = np.zeros(3)
b1
```

```python
y1 = sigmoid(X@W1 + b1)
y1
```

```python
W2 = np.random.randn(3,3)/100
W2.round(3)
```

```python
b2 = np.zeros(3)
b2
```

```python
y2 = sigmoid(y1@W2 + b2)
y2
```
